```python
import matplotlib.pyplot as plt
import numpy as np 
import os

d = './'

for p in range(6):
    if(p == 0):
        continue
    else:
        points1 = []
        points2 = []
        s = 'test' + str(p) + '.txt'
        for filename in os.listdir(d):
            if not filename.endswith(s):
                continue

            with open(filename, 'r') as f:
                first = 1
                for line in f.readlines():
                    if first:
                        first = 0
                        continue
                    else:
                        point = (line.split(' '))
                        x1 = float(point[0])
                        y1 = float(point[1])
                        x2 = float(point[2])
                        y2 = float(point[3])
                        
                        points1.append([x1, x2])
                        points2.append([y1, y2])

        points1 = np.array(points1)
        # print(points1)
        points2 = np.array(points2)
        # print(points2)
        
        plt.figure(facecolor='#00517c')

        ax = plt.axes()
        ax.spines["bottom"].set_color("white")
        ax.spines["top"].set_color("white")
        ax.spines["right"].set_color("white")
        ax.spines["left"].set_color("white")
        ax.set_facecolor("#00517c")

        ax.tick_params(axis='x', colors='white')
        ax.tick_params(axis='y', colors='white')

        for i in range(len(points1)):
            plt.plot(points1[i], points2[i], 'k-', color="white")


        points = []

        s1 = 'output' + str(p) + '.txt'
        for filename in os.listdir(d):
            if not filename.endswith(s1):
                continue

            with open(filename, 'r') as f:
                for line in f.readlines():
                    point = (line.split(', '))
                    x = float(point[0])
                    y = float(point[1])

                    points.append([x, y])

        points = np.array(points)

        plt.plot(points[:, 0], points[:, 1], 'o', color="red")
        # naming the x axis 
        plt.xlabel('x - axis', color="white") 
        # naming the y axis 
        plt.ylabel('y - axis', color="white")

        plt.show()
```
