```python
import os
import numpy as np

d = './'

points = []

for filename in os.listdir(d):
    if not filename.endswith('input.csv'):
        continue

    with open(filename, 'r') as f:
        first = 1
        for line in f.readlines():
            if first:
                first = 0
                continue
            else:
                point = (line.split(', '))
                x = int(point[0])
                y = int(point[1])
                points.append([x, y])

points = np.array(points)
# print(points)
```

```python
import matplotlib.pyplot as plt

plt.figure(facecolor='#00517c')

ax = plt.axes()
ax.spines["bottom"].set_color("white")
ax.spines["top"].set_color("white")
ax.spines["right"].set_color("white")
ax.spines["left"].set_color("white")
ax.set_facecolor("#00517c")

ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

plt.plot(points[:, 0], points[:, 1], 'o', color="white")
d = './'

lines = []
seperation_line = []

for filename in os.listdir(d):
    if not filename.endswith('output.txt'):
        continue

    with open(filename, 'r') as f:
        for line in f.readlines():
            if line.find(', ') == -1: 
                point = (line.split(' '))
                x = int(point[0])
                y = int(point[1])
                seperation_line.append([x, y])
            else:
                point = (line.split(', '))
                x = int(point[0])
                y = int(point[1])
                lines.append([x, y])

seperation_line = np.array(seperation_line)
lines = np.array(lines)

n = len(lines)

for i in range(n):
    plt.plot([lines[i][0], lines[(i + 1) % n][0]], [
             lines[i][1], lines[(i + 1) % n][1]], 'white')

plt.plot([seperation_line[0][0], seperation_line[1][0]], [seperation_line[0][1], seperation_line[1][1]], 'white')


plt.xlabel('x - axis', color="white")
# # naming the y axis
plt.ylabel('y - axis', color="white")
plt.show()
```
