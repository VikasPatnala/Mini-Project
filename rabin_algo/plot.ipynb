```python
import os
import numpy as np

d = './'

points = []

for filename in os.listdir(d):
    if not filename.endswith('input.csv'):
        continue

    with open(filename, 'r') as f:
        first = 1
        for line in f.readlines():
            if first:
                first = 0
                continue
            else:
                point = (line.split(', '))
                x = int(point[0])
                y = int(point[1])
                points.append([x, y])

points = np.array(points)
# print(points)

arr = []

for filename in os.listdir(d):
    if not filename.endswith('output.txt'):
        continue

    with open(filename, 'r') as f:
        for line in f.readlines():
            arr_pt = (line.split(' '))
            x = int(arr_pt[0])
            y = int(arr_pt[1])
            arr.append([x, y])

arr = np.array(arr)

print(arr)
```
