```python
import matplotlib.pyplot as plt
import numpy as np
import math

import os

for i in range(3):
    d = './'

    points = []

    reg = []
    if i != 0:
        s = 'input' + str(i) + ".csv"
        for filename in os.listdir(d):
            if not filename.endswith(s):
                continue
            with open(filename, 'r') as f:
                for line in f.readlines():
                    if line.find(', ') == -1:
                        if line.find('reg') != -1:
                            point = (line.split(' '))
                            reg.append(int(point[1]))
                            reg.append(int(point[2]))
                            reg.append(int(point[3]))
                            reg.append(int(point[4]))
                    else:
                        point = (line.split(', '))
                        x = int(point[0])
                        y = int(point[1])
                        points.append([x, y])

        reg = np.array(reg)
        points = np.array(points)


        lines = []
        qpoints = []

        s = 'output' + str(i) + ".txt"
        for filename in os.listdir(d):
            if not filename.endswith(s):
                continue
            with open(filename, 'r') as f:
                for line in f.readlines():
                    if line.find(', ') == -1:
                        point = (line.split(' '))
                        axis = (point[0])
                        value = int(point[1])
                        lines.append([axis, value])
                    else:
                        point = (line.split(', '))
                        x = int(point[0])
                        y = int(point[1])
                        qpoints.append([x, y])

        lines = np.array(lines)
        qpoints = np.array(qpoints)
        plt.figure(facecolor='#00517c')

        ax = plt.axes()
        ax.spines["bottom"].set_color("white")
        ax.spines["top"].set_color("white")
        ax.spines["right"].set_color("white")
        ax.spines["left"].set_color("white")
        ax.set_facecolor("#00517c")

        ax.tick_params(axis='x', colors='white')
        ax.tick_params(axis='y', colors='white')
        # plt.plot([69, 69], [-20, 120], color='white')
        for i in range(len(lines)):
            if lines[i][0] == 'x':
                plt.plot([int(lines[i][1]), int(lines[i][1])], [-20, 120], color='white')
            elif lines[i][0] == 'y':
                plt.plot([-20, 120], [int(lines[i][1]), int(lines[i][1])], color='yellow')

        plt.xlim(0, 100)
        plt.ylim(0, 100)

        plt.plot([reg[0], reg[1]], [reg[2], reg[2]], color='red')
        plt.plot([reg[0], reg[1]], [reg[3], reg[3]], color='red')
        plt.plot([reg[0], reg[0]], [reg[2], reg[3]], color='red')
        plt.plot([reg[1], reg[1]], [reg[2], reg[3]], color='red')

        # naming the x axis
        plt.xlabel('x - axis', color="white")
        # naming the y axis
        plt.ylabel('y - axis', color="white")

        plt.plot(points[:, 0], points[:, 1], 'o', color="white")
        plt.plot(qpoints[:, 0], qpoints[:, 1], 'o', color="red")
```
