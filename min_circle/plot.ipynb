```python
import matplotlib.pyplot as plt
import numpy as np
import math

import os

d = './'

points = []

plt.figure(facecolor='#00517c')

ax = plt.axes()
ax.spines["bottom"].set_color("white")
ax.spines["top"].set_color("white")
ax.spines["right"].set_color("white")
ax.spines["left"].set_color("white")
ax.set_facecolor("#00517c")

ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

for filename in os.listdir(d):
    if not filename.endswith('input.csv'):
        continue
    with open(filename, 'r') as f:
        for line in f.readlines():
            if line.find(',') == -1:
                x = 0
            else:
                point = (line.split(','))
                x = int(point[0])
                y = int(point[1])
                points.append([x, y])

points = np.array(points)


# naming the x axis
plt.xlabel('x - axis', color="white")
# naming the y axis
plt.ylabel('y - axis', color="white")

radius = 0
center_x = 0
center_y = 0

for filename in os.listdir(d):
    if not filename.endswith('output.txt'):
        continue
    with open(filename, 'r') as f:
        for line in f.readlines():
            if line.find(',') == -1:
                radius = float(line)
            else:
                point = (line.split(','))
                center_x = float(point[0])
                center_y = float(point[1])

# print(radius)
# print(center_x)
# print(center_y)

plt.plot(points[:, 0], points[:, 1], 'o', color="white")

circle = plt.Circle((center_x, center_y), radius, color='red', fill=False)

plt.plot(center_x, center_y, 'o', color='red')

fig = plt.gcf()
ax = plt.gca()

ax.set_aspect(1)
ax.add_artist(circle)

ax.add_patch(circle)
plt.show()
```
