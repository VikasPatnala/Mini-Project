```python
import matplotlib.pyplot as plt
import numpy as np
import math

import os

d = './'

points1 = []
points2 = []

plt.figure(facecolor='#00517c')

ax = plt.axes()
ax.spines["bottom"].set_color("white")
ax.spines["top"].set_color("white")
ax.spines["right"].set_color("white")
ax.spines["left"].set_color("white")
ax.set_facecolor("#00517c")

ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

for filename in os.listdir(d):
    if not filename.endswith('input.csv'):
        continue
    x = 0
    with open(filename, 'r') as f:
        for line in f.readlines():
            if line.find(',') == -1:
                x += 1
                if x == 3:
                    break
            else:
                point = (line.split(','))
                if x == 1:
                    x1 = float(point[0])
                    y1 = float(point[1])
                    points1.append([x1, y1])
                elif x == 2:
                    x2 = float(point[0])
                    y2 = float(point[1])
                    points2.append([x2, y2])

points1 = np.array(points1)
points2 = np.array(points2)


# naming the x axis
plt.xlabel('x - axis', color="white")
# naming the y axis
plt.ylabel('y - axis', color="white")


plt.plot(points1[:, 0], points1[:, 1], 'o', color="white")
plt.plot(points2[:, 0], points2[:, 1], 'o', color="yellow")


def distance(x, y, x1, y1):
  return math.sqrt((x - x1) ** 2 + (y - y1) ** 2)


for i in range(len(points1)):
  for j in range(len(points2)):
    if distance(points1[i][0], points1[i][1], points2[j][0], points2[j][1]) <= 4:
      plt.plot([points1[i][0], points2[j][0]], [
               points1[i][1], points2[j][1]], color="white")

res1 = []
res2 = []

for filename in os.listdir(d):
    if not filename.endswith('output.txt'):
        continue

    with open(filename, 'r') as f:
        for line in f.readlines():
            point = (line.split(' '))
            x1 = float(point[0])
            y1 = float(point[1])
            x2 = float(point[2])
            y2 = float(point[3])

            res1.append([x1, x2])
            res2.append([y1, y2])

res1 = np.array(res1)
res2 = np.array(res2)

for i in range(len(res1)):
    plt.plot(res1[i], res2[i], color="red")
```
